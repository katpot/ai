##c. Uniform Cost Search

```python
#Using classes
class Node:
    def __init__(self, state, parent=None, cost=0):
        self.state = state
        self.parent = parent
        self.cost = cost

def sort(frontier):
    n = len(frontier)
    for i in range(n):
        min = i
        for j in range(i + 1, n):
            if frontier[j].cost < frontier[min].cost:
                min = j
        frontier[i], frontier[min] = frontier[min], frontier[i]

class UniformCostSearch:
    def __init__(self, graph, start, goal):
        self.graph = graph
        self.start = start
        self.goal = goal
        self.explored = set()
        self.frontier = [Node(start, None, 0)]

    def search(self):
        while self.frontier:
            sort(self.frontier)
            node = self.frontier.pop(0)
            self.explored.add(node.state)
            print("F:", [node.state for node in self.frontier])
            print("E:",self.explored)
            print(f"Node: {node.state}  Parent: {node.parent}  Cost: {node.cost}\n\n")

            if node.state == self.goal:
                print("\nGoal node found")
                return

            for neighbor, cost in self.graph[node.state].items():
                if neighbor not in self.explored:
                    neighbor_node = Node(neighbor, node.state, node.cost + cost)
                    self.frontier.append(neighbor_node)

        print("Goal node not found")

if __name__ == '__main__':
    graph = eval(input("Enter graph: "))
    start = input("Enter start node: ")
    goal = input("Enter goal node: ")
    ucs = UniformCostSearch(graph, start, goal)
    ucs.search()
```

```python
class Graph:
    def __init__(self, graph):
        self.graph = graph

    def reverse(self):
        reversed_graph = {}
        for node in self.graph:
            reversed_graph[node] = []
        for node, neighbors in self.graph.items():
            for neighbor in neighbors:
                reversed_graph[neighbor].append(node)
        return reversed_graph

class BiDirectionalSearch:
    def __init__(self, graph):
        self.graph = graph

    def search(self, start, goal):
        rev_graph = Graph(self.graph).reverse()

        explored1 = []
        explored2 = []
        frontier1 = []
        frontier2 = []
        frontier1.append([start, None])
        frontier2.append([goal, None])
        path1 = []
        path2 = []
        intersect = None

        while frontier1 and frontier2:
            # ------------------------------------------------------------------Forward search
            print("F_Front ", frontier1)
            print("E_Front ", explored1)
            node, parent = frontier1.pop()
            path1.append([node, parent])

            if node in explored2:
                intersect = node
                break

            if node not in explored1:
                explored1.append(node)

                for neighbor in self.graph[node]:
                    if neighbor not in explored1:
                        frontier1.append([neighbor, node])

            # -------------------------------------------------------------------Backward search
            print("F_Back ", frontier2)
            print("E_Back ", explored2, "\n")
            node2, parent2 = frontier2.pop(0)
            path2.append([node2, parent2])

            if node2 in explored1:
                intersect = node2
                break

            if node2 not in explored2:
                explored2.append(node2)

                for neighbor in rev_graph[node2]:
                    if neighbor not in explored2:
                        frontier2.append([neighbor, node2])

        path2 = [[node2, node1] for node1, node2 in path2]
        print("The Path from ",start," to ",goal," is: ",path1+path2[::-2])
        print("The search intersected at: ", intersect)


if __name__ == '__main__':
    graph = eval(input("Enter the graph: "))
    bds = BiDirectionalSearch(graph)
    s = input("Enter Start node: ")
    g = input("Enter goal node: ")
    print("\n")
    bds.search(s, g)
```

#2. AIM:Implement water jug problem with Search tree generation

```python
path = []

class WaterJugProblem:
    def __init__(self, b, s, t):
        self.b = b
        self.s = s
        self.t = t

    def fbj(self, frontier, explored, big, small):
        if (self.b, small) not in frontier and (self.b, small) not in explored:
            frontier.append((self.b, small))

    def fsj(self, frontier,  explored, big, small):
        if (big, self.s) not in frontier and (big, self.s) not in explored:
            frontier.append((big, self.s))

    def ebj(self, frontier, explored, big, small):
        if (0, small) not in frontier and (0, small) not in explored:
            frontier.append((0, small))

    def esj(self, frontier,  explored, big, small):
        if (big, 0) not in frontier and (big, 0) not in explored:
            frontier.append((big, 0))

    def pbs(self, frontier, explored,  big, small):
        transfer = min(big, self.s - small)
        if (big - transfer, small + transfer) not in frontier and (big - transfer, small + transfer) not in explored:
            frontier.append((big - transfer, small + transfer))

    def psb(self, frontier, explored,  big, small):
        transfer = min(small, self.b - big)
        if (big + transfer, small - transfer) not in frontier and (big + transfer, small - transfer) not in explored:
            frontier.append((big + transfer, small - transfer))

    def bfs(self):
        explored = set()
        frontier = [(0, 0)]

        while frontier:
            print('f:', frontier, '\ne:', explored, '\n')
            big, small = frontier.pop(0)

            if big == self.t or small == self.t:
                print("t achieved:", path + [(big, small)])
                return

            if (big, small) not in explored and (big, small) not in frontier:
                explored.add((big, small))
                path.append((big, small))

                self.fbj(frontier, explored, big, small)
                self.fsj(frontier,  explored, big, small)
                self.ebj(frontier,  explored, big, small)
                self.esj(frontier,  explored, big, small)
                self.pbs(frontier,  explored, big, small)
                self.psb(frontier,  explored, big, small)

        print("t not achievable.")


if __name__ == '__main__':
    b = int(input("Enter Big jug capacity: "))
    s = int(input("Enter Small jug capacity: "))
    t = int(input("Enter t jug capacity: "))

    WaterJugProblem(b, s, t).bfs()
```

#3. AIM:Implement Missionaries and Cannibals problem with Search tree generation

```python
class Node:
    def __init__(self, name, parent, heuristic):
        self.name = name
        self.parent = parent
        self.heuristic = heuristic

class GraphSearch:
    def __init__(self, graph):
        self.graph = graph
        self.explored = []
        self.frontier = []

    def sort(self):
        n = len(self.frontier)
        for i in range(n):
            min = i
            for j in range(i + 1, n):
                if self.frontier[j].heuristic < self.frontier[min].heuristic:
                    min = j
            self.frontier[i], self.frontier[min] = self.frontier[min], self.frontier[i]

    def best_first_search(self, start, goal):
        self.frontier.append(Node(start, None, self.graph[start]['X']))

        while self.frontier:
            self.sort()
            cur_node = self.frontier.pop(0)
            print("F:", [node.name for node in self.frontier])
            print("E:",self.explored)
            print("Node: ", cur_node.name, "  Parent: ", cur_node.parent, "  Heuristic: ", cur_node.heuristic,"\n\n")

            if cur_node.name == goal:
                print("\nGoal node found")
                break

            if cur_node.name not in self.explored:
                self.explored.append(cur_node.name)

                for neighbor in self.graph[cur_node.name]:
                    if neighbor != 'X' and neighbor not in self.explored:
                        neighbor_heuristic = self.graph[neighbor]['X']
                        self.frontier.append(Node(neighbor, cur_node.name, neighbor_heuristic))

        if cur_node.name != goal:
            print("Goal node not found")

if __name__ == '__main__':

    graph = eval(input("Enter graph: "))
    search = GraphSearch(graph)
    start = input("Enter start node:")
    goal = input("Enter goal node:")
    search.best_first_search(start,goal)
```

#6. Implement 8-puzzle problem using A* algorithm.

```python
def sort_frontier(frontier):
    n=len(frontier)
    for i in range(n):
        for j in range(n):
            if frontier[i][1]<frontier[j][1]:
                frontier[i],frontier[j] = frontier[j],frontier[i]

def heuristic(start, goal):
    count = 0
    for i in range(len(start)):
        if start[i] != goal[i]:
            count += 1
    return count

def up(frontier, explored, state, goal):
    zero = state.index(0)
    if zero not in [0, 1, 2]:
        new_state = state[:]
        new_state[zero], new_state[zero - 3] = new_state[zero - 3], new_state[zero]
        if new_state not in [item[0] for item in frontier] and new_state not in explored:
            frontier.append([new_state, heuristic(new_state, goal)])

def down(frontier, explored, state, goal):
    zero = state.index(0)
    if zero not in [6, 7, 8]:
        new_state = state[:]
        new_state[zero], new_state[zero + 3] = new_state[zero + 3], new_state[zero]
        if new_state not in [item[0] for item in frontier] and new_state not in explored:
            frontier.append([new_state, heuristic(new_state, goal)])

def left(frontier, explored, state, goal):
    zero = state.index(0)
    if zero not in [0, 3, 6]:
        new_state = state[:]
        new_state[zero], new_state[zero - 1] = new_state[zero - 1], new_state[zero]
        if new_state not in [item[0] for item in frontier] and new_state not in explored:
            frontier.append([new_state, heuristic(new_state, goal)])

def right(frontier, explored, state, goal):
    zero = state.index(0)
    if zero not in [2, 5, 8]:
        new_state = state[:]
        new_state[zero], new_state[zero + 1] = new_state[zero + 1], new_state[zero]
        if new_state not in [item[0] for item in frontier] and new_state not in explored:
            frontier.append([new_state, heuristic(new_state, goal)])

def a_star_8queen(start, goal):
    frontier = []
    explored = []
    frontier.append([start, heuristic(start, goal)])

    while frontier:
        print("Frontier: ", frontier)
        print("Explored: ", explored)
        sort_frontier(frontier)
        state, h = frontier.pop(0)

        print("State: ", state, " Heuristic: ", h,'\n\n')

        if state == goal:
            print("Goal state achieved: ", goal)
            return

        if state not in explored:
            explored.append(state)
            up(frontier, explored, state, goal)
            down(frontier, explored, state, goal)
            left(frontier, explored, state, goal)
            right(frontier, explored, state, goal)

if __name__ == '__main__':
    start = eval(input("Enter start state: "))
    goal = eval(input("Enter goal state: "))
    a_star_8queen(start, goal)
```

#7. Implement AO* algorithm for General graph problem.

```python
def Cost(H, condition, weight=1):
    cost = {}
    if 'AND' in condition:
        AND_nodes = condition['AND']
        Path_A = ' AND '.join(AND_nodes)
        PathA = sum(H[node] + weight for node in AND_nodes)
        cost[Path_A] = PathA
    if 'OR' in condition:
        OR_nodes = condition['OR']
        Path_B = ' OR '.join(OR_nodes)
        PathB = min(H[node] + weight for node in OR_nodes)
        cost[Path_B] = PathB
    return cost

def update_cost(H, Conditions, weight=1):
    Main_nodes = list(Conditions.keys())
    Main_nodes.reverse()
    least_cost = {}
    for key in Main_nodes:
        condition = Conditions[key]
        print(key, ':', Conditions[key], '>>>', Cost(H, condition, weight))
        c = Cost(H, condition, weight)
        H[key] = min(c.values())
        least_cost[key] = Cost(H, condition, weight)
    return least_cost

def shortest_path(Start, Updated_cost, H):
    Path = Start
    if Start in Updated_cost.keys():
        Min_cost = min(Updated_cost[Start].values())
        key = list(Updated_cost[Start].keys())
        values = list(Updated_cost[Start].values())
        Index = values.index(Min_cost)
        Next = key[Index].split()
        if len(Next) == 1:
            Start = Next[0]
            Path += '<--' + shortest_path(Start, Updated_cost, H)
        else:
            Path += '<--(' + key[Index] + ') '
            Start = Next[0]
            Path += '[' + shortest_path(Start, Updated_cost, H) + ' + '
            Start = Next[-1]
            Path += shortest_path(Start, Updated_cost, H) + ']'
    return Path

H = eval(input('Enter nodes with heuristic costs: '))
Conditions = eval(input('Enter graph: '))
weight = 1
print('Updated Cost:')
Updated_cost = update_cost(H, Conditions, weight=1)
print('Shortest Path:\n', shortest_path('A', Updated_cost, H))
```

##a. MINMAX Algorithm

```python
import math

def minimax(game, state):
    def maxvalue(game, state):
        if state not in game:
            return state
        v = -math.inf
        for a in game[state]:
            v=max(v,minvalue(game,a))
        print("Max value for ",state," : ",v)
        return v

    def minvalue(game,state):
        if state not in game:
            return state
        v = math.inf
        for a in game[state]:
            v = min(v,maxvalue(game,a))
        print("Min value for ",state," : ",v)
        return v

    return max([minvalue(game,a) for a in game[state]])

graph = eval(input("Enter graph: "))
ans = minimax(graph,'A')
print("The optimal value is: ",ans)
```

##b. Alpha-Beta Pruning

```python
import math

def alphaBeta(game, state):
    def maxvalue(game, state,alpha,beta):
        if state not in game:
            return state
        v = -math.inf
        for a in game[state]:
            v=max(v,minvalue(game,a,alpha,beta))
            if v >= beta:
                print("Pruned subtree at state:", state, "with value:", v)
                return v
            alpha = max(alpha,v)
        print("State: ",state," V: ",v," Alpha: ",alpha," Beta: ",beta)
        return v

    def minvalue(game,state,alpha,beta):
        if state not in game:
            return state
        v = math.inf
        for a in game[state]:
            v=min(v,maxvalue(game,a,alpha,beta))
            if v <= alpha:
                print("Pruned subtree at state:", state, "with value:", v)
                return v
            beta = min(beta,v)
        print("State: ",state," V: ",v," Alpha: ",alpha," Beta: ",beta)
        return v

    alpha = -math.inf
    beta = math.inf
    v = maxvalue(game,state,alpha,beta)
    return v

graph = eval(input("Enter graph: "))
ans = alphaBeta(graph,'A')
print("The optimal value is: ",ans)
```

#9. Implement Cryptartihmetic problems

```python
import itertools

def number(n, d):
    t = 0
    for i in n:
        t = d[i] + (t * 10)
    return t

def test(l, s, d):
    total = 0
    for i in l:
        total += number(i, d)
    return total == number(s, d)

def check(d, c):
    for i in d.keys():
        if i in c and d[i] == 0:
            return True
    return False

def solve(l, s):
    c = [i[0] for i in l] + [s[0]]
    p = list(set(''.join(l) + s))
    q = len(p)
    permutations = itertools.permutations(range(0, 10), q)
    for perm in permutations:
        d = {p[j]: perm[j] for j in range(q)}
        if check(d, c):
            continue
        if test(l, s, d):
            return d
    return None

if __name__ == "__main__":
    l = input('Enter list of strings: ').split()
    s = input('Enter output string: ')
    solution = solve(l, s)
    if solution:
        print(solution)
        print('Solution found')
    else:
        print('No solution found')
```
